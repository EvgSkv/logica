# Example of List Comprehension in Logica

```python
from logica import colab_logica 
colab_logica.SetTabulatedOutput(False)
```

```python
%%logica PremiumPurchases

@Engine("sqlite");

Purchases(customer: "John",                                                                                                                          
          tickets: [{price: 10, tier: "economy"},{price: 20, tier: "premium"}]);                                                                     
Purchases(customer: "James",                                                                                                                         
          tickets: [{price: 5, tier: "economy"}]);                                                                                                   
                                                                                                                                                     
PremiumPurchases(customer:, premium_tickets:, premium_total:) :-
  Purchases(customer:, tickets:),                                                                                                                  
  premium_tickets List= (                                                                                                                          
    ticket :- ticket in tickets, ticket.tier == "premium"                                                                                          
  ),                                                                                                                                               
  premium_total += (ticket.price :- ticket in premium_tickets),                                                                                    
  Size(premium_tickets) > 0;    
```
