## Calculating distribution of customers by order count

```python
%%logica CustomerDistribution

@Engine("sqlite");
@AttachDatabase("db", "tpc.db");

OrdersByCustomer(c_custkey) += 1 :-
  db.customer(c_custkey:),
  db.orders(o_custkey:, o_comment:),
  o_custkey == c_custkey,
  !Like(o_comment, "%special%requesets%");
OrdersByCustomer(c_custkey) += 0 :-
  db.customer(c_custkey:);

CustomerDistribution(
  c_count: OrdersByCustomer(),
  custdist? += 1) distinct;
```
