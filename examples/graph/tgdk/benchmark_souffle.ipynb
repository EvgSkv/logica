```python
graphs = ['g1k', 'g2k', 'g3k', 'g4k', 'g5k']
```

```python
%%loop ('writing TC programs (no printing)', graphs)
%%writefile tc_for_{loop_parameter}.s

.decl g(a:number,b:number)
.decl tc(a:number, b:number)
.decl result(a:number, b:number)

.input g(filename="data/{loop_parameter}.csv",delimiter=",",headers=true)

tc(a, b) :- g(a, b).
tc(a, c) :- tc(a, b), g(b, c).

result(a, b) :- tc(a, b).

.output result(filename="output/tc_{loop_parameter}.csv",delimiter=",")
```

```python
trees = ['tree7', 'tree8', 'tree9', 'tree10', 'tree11', 'tree12']
```

```python
%%loop ('writing SG programs (no printing)', trees)
%%writefile sg_for_{loop_parameter}.s

.decl g(a:number,b:number)
.decl sg(a:number, b:number)
.decl result(a:number, b:number)

.input g(filename="data/{loop_parameter}.csv",delimiter=",",headers=true)

sg(a, b) :- g(x, a), g(x, b).
sg(a, b) :- sg(x, y), g(x, a), g(y, b).

result(a, b) :- sg(a, b).

.output result(filename="output/sg_{loop_parameter}.csv",delimiter=",")
```

```python
%%loop ('Souffle SG (no printing)', trees)
!echo 'Compiling SG for {loop_parameter}'
!souffle -o sg_for_{loop_parameter} sg_for_{loop_parameter}.s --jobs=32
```

```python
%%loop ('Souffle SG', trees)
!echo 'Running SG for {loop_parameter}'
!./sg_for_{loop_parameter}  --jobs=32
```
